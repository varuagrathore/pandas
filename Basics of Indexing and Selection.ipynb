**Multi-Indexing and Hierarchical Indexing**

```python
#this help when we use multilevel index for example
df_new=data.set_index(['age','card'])
```

```python
df_new
```

```python
#now filter based on this 
df_new.loc[(45,'yes')]
```

```python
#assume when you deal with sales analysis and you need to divide in group and subgroup then this help there 
```

```python
#to swap level
data_swap = df_new.swaplevel('age','card')
```

```python
data_swap
```

```python
#now reset index
data_flat = data_swap.reset_index()
```

```python
data_flat
```

```python
#Selecting Data at Specific Level
# You can use the xs() method to select data at a specific level. For example,
 #to select all data where 'age' is '39':
```

```python
df_new.xs(39,level='age',axis=0)
```

multi-indexing or hierarchical indexing in Pandas allows you
 to have multiple index levels within a single DataFrame, providing a more
 structured and intuitive way to handle higher-dimensional data. By creating
a multi-level index with the set_index() method and using various methods
 like loc[], xs(), swaplevel(), and sort_index(), you can efficiently
 manipulate and access data within a multi-level index.


**Indexing with Booleans and Conditional Selection**

```python
#These techniques are powerful tools for data
 #manipulation and can be applied in various contexts, such as data
 #preprocessing, exploratory data analysis, and machine learning.
ans = (data['age']<=39)
ans
```

```python
#now apply boolean mask to the dataframe
#this mask is applied to the DataFrame, only the rows with True values are selected.
true_only = data[data['age']==39]
```

```python
true_only
```

```python
#we can apply multiple condition
mul_condition = data[(data['card']=='yes') & (data['owner']=='no')]
```

```python
mul_condition
```

```python
#we can apply multiple condition
mul_condition_or = data[(data['card']=='yes') | (data['owner']=='no')]
```

```python
mul_condition_or
```

```python
#we can also used query method
query_selection = data.query(" age==45 and card =='yes'")
```

```python
query_selection
```

Boolean indexing and conditional selection in Pandas allow
 you to filter rows in a DataFrame based on specific conditions. By creating
 boolean masks, combining multiple conditions with & or | operators, or
 using the query() method, you can perform various data analysis, cleaning,
 and transformation tasks more efficiently.
