```python
import pandas as pd
```

```python
#split the data into two subsets
data = pd.read_csv(r"C:\Users\dell\Downloads\archive\AER_credit_card_data.csv")
```

```python
data
```

```python
data_yes = data[data['card'] == 'yes']
data_no = data[data['card'] == 'no']
```

```python
data_yes.reset_index()
```

```python
data_no.reset_index()
```

```python
#concatenate the DataFrame along the rows (axis=0)
concatenated_data = pd.concat([data_yes,data_no])
```

```python
concatenated_data
```

```python
concatenated_data1 = pd.concat([data_yes,data_no],axis=1)
concatenated_data1
```

Keep in mind that when concatenating DataFrames along columns, it's
 crucial that they have the same index. Otherwise, you might end up with
 missing values in the result.
