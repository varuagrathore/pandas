 This technique proves to be useful when analyzing data based on specific
 categories or combinations of categories. Through the "groupby()" function,
 it is possible to organize data into groups that have similar values and
 perform mathematical operations on the data within those groups. 

```python
import pandas as pd
```

```python
#split the data into two subsets
data = pd.read_csv(r"C:\Users\dell\Downloads\archive\AER_credit_card_data.csv")
data
```

```python
grouped_data = data.groupby(['card','owner'])
```

```python
print(grouped_data)
```

```python
# You can now apply aggregation functions to the grouped data.
grouped_data['age'].mean()
```

```python
grouped_data['age'].agg(['mean',
 'min', 'max'])
```
