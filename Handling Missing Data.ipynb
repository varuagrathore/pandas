**Fill Missing values**

The fillna() methods fills missing values with a specified value or using a specified method( such as forward fill or backward fill)

```python
#To fill missing value with the mean of the respective column:
data_filled0 = data.fillna(0)
```

```python
data_filled0
```

```python
#To fill missing values with the mean of the respective column:
data_filled = data.fillna(data.mean())
```

```python
data_filled
```

```python
data.describe().T
```

```python
# To fill missing values using forward fill (propagate the last valid observation forward):
data_filled1 = data.ffill()
```

```python
data_filled1
```

```python
data_filled2 = data.bfill()
```

```python
data_filled2.isnull().sum()
```

```python
data_filled3 = data.fillna(data.median())
```

```python
data_filled3
```

**Mean:** Sensitive to outliers and skewed data. A single extreme value can significantly affect the mean, making it unrepresentative of the central tendency of the majority of the data.

**Median:** Not affected by outliers or skewed data. The median is the middle value when data is sorted and provides a better central tendency measure in the presence of outliers.

Pandas provides several methods to handle missing data in
DataFrames. By using methods like dropna() and fillna(), you can
efficiently handle missing values based on your specific requirements and
the nature of the data
